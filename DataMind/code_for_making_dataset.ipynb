```python
def movie_info(id):
    response2=requests.get('https://api.themoviedb.org/3/movie/'+str(id)+'/videos?api_key=08daf48277f2de509dbcd4e840fb26d8')
    res2=response2.json()
    trailer=[]
    for i in res2['results']:
        if(i['official']==True):
            trailer.append('https://www.youtube.com/watch?v='+i['key'])
            break
    if(len(trailer)==0):
        df1=pd.DataFrame()
        return df1
    else:
        trailer=trailer[0]
        
    response=requests.get('https://api.themoviedb.org/3/movie/'+str(id)+'?api_key=08daf48277f2de509dbcd4e840fb26d8')
    res=response.json()
    
    movie_id=res['id']
    movie_title=res['title']
    overview=res['overview']
    genre=res['genres']
    release_date=res['release_date']
    runtime=res['runtime']
    tagline=res['tagline']
    if res['poster_path']!=None:
        poster_image='https://image.tmdb.org/t/p/w500'+res['poster_path']
    else:
        poster_image=None

    if res['backdrop_path']!=None:
        background_image='https://image.tmdb.org/t/p/w500'+res['backdrop_path']
    else:
        background_image=None
   
    if((res['spoken_languages']!=None) & (len(res['spoken_languages'])>0)):
        language=res['spoken_languages'][0]['english_name']
    else:
        language=None

    if((res['production_countries']!=None) & (len(res['production_countries'])>0)):
        country=res['production_countries'][0]['name']
    else:
        country=None
    
    rating=res['vote_average']
    A_rated=res['adult']

    response1=requests.get('https://api.themoviedb.org/3/movie/'+str(id)+'/credits?api_key=08daf48277f2de509dbcd4e840fb26d8')
    res1=response1.json()
    cast=[]
    cnt=0
    for i in res1['cast']:
        if(i['known_for_department']=='Acting'):
            if(i['profile_path']!=None):
                if(cnt<10):
                    cnt=cnt+1
                    cast.append({"actor_id":i['id'], "name":i['original_name'], "character":i['character'], "image":'https://image.tmdb.org/t/p/w500'+i['profile_path']})
    director=[]
    writer=[]
    for i in res1['crew']:
        if(i['job']=='Director'):
            director.append({"director_id":i['id'], "name":i['original_name']})
        elif(i['department']=='Writing'):
                writer.append(i['original_name'])

    if(len(director)==0):
        df1=pd.DataFrame()
        return df1
    else:
        director=director[0]
        
    if(len(writer)==0):
        writer=None
    else:
        writer=writer[0]
        

    data=[{'movie_id':movie_id,
          'movie_title':movie_title,
          'overview':overview,
          'genre':genre,
          'release_date':release_date,
          'runtime':runtime,
          'Director':director,
          'Writer':writer,
          'Cast':cast,
          'tagline':tagline,
          'poster_image':poster_image,
          'background_image':background_image,
          'language':language,
          'country':country,
          'rating':rating,
          'A-rated':A_rated,
          'trailer':trailer,
          'count':0}]
    df=pd.DataFrame(data)
    return df
```
