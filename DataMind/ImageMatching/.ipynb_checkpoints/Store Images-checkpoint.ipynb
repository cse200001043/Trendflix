```python
import json


def extract_image_urls(jsonl_file_path):
    # Create an empty list to store the extracted data
    data_list = []

    # Open the JSONL file and process each line
    with open(jsonl_file_path, 'r') as file:
        for line in file:
            # Parse each line as JSON
            try:
                data = json.loads(line)
                data_list.append(data)
            except json.JSONDecodeError:
                print(f"Error parsing JSON in line: {line}")

    # Now, data_list contains the extracted data from the JSONL file
    # You can access and manipulate the data as needed

    return data_list
#     image_urls = []
#     # For example, to print the extracted data
#     for item in data_list:
#         image_urls.append(item['img_src'])
        
#     return image_urls


file_image_data = extract_image_urls("../Data/top.jsonl")
file_image_data
```

```python
import requests
from PIL import Image
from io import BytesIO


def download_image_from_url(image_url, local_file_path):
    # Send an HTTP GET request to the URL
    response = requests.get(image_url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Read the image content from the response
        image_data = response.content

        # Create an image object from the image data
        image = Image.open(BytesIO(image_data))

        # Specify the local file path to save the image

        # Save the image to the local file system
        image.save(local_file_path)
        print(f"Image downloaded and saved as {local_file_path}")
    else:
        print(f"Failed to download the image. Status code: {response.status_code}")
        
        
for iter in range(len(file_image_data)):
    url = file_image_data[iter]['img_src'][0]
    download_image_from_url(url, "Images/" + str(iter) + ".jpg")
```

#  SIFT Algorithm ChatGPT implementation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import pairwise_distances_argmin_min

# Load the query image
query_image_path = 'Images/0.jpg'
query_image = cv2.imread(query_image_path, 0)  # Load the image in grayscale

# Directory containing images in your database
database_directory = 'Images/'

# Initialize lists to store image paths and descriptors
image_paths = []
descriptors = []

# Load and compute descriptors for images in the database
for filename in os.listdir(database_directory):
    if filename.endswith('.jpg'):
        image_path = os.path.join(database_directory, filename)
        image = cv2.imread(image_path, 0)  # Load the image in grayscale
        image_paths.append(image_path)
        
#         plt.imshow(image)
#         plt.axis('off')  # Turn off the axis labels
#         plt.show()
        
        # Compute image descriptor (e.g., HOG)
        hog = cv2.HOGDescriptor()
        descriptor = hog.compute(image)
        descriptors.append(descriptor)

# Compute descriptors for the query image
hog = cv2.HOGDescriptor()
query_descriptor = hog.compute(query_image)
```

```python
# Convert descriptors to a NumPy array for easier processing
descriptors = np.array(descriptors)

print(len(descriptors[0]))

query_descriptor = query_descriptor.reshape(1, -1)
# Compute distances between the query descriptor and all database descriptors
distances = pairwise_distances_argmin_min(query_descriptor, descriptors)

# # Find the index of the image with the minimum distance
# most_similar_index = distances[0][0]

# # Get the path of the most similar image
# most_similar_image_path = image_paths[most_similar_index]

# print(f"The most similar image is: {most_similar_image_path}")
```

# SIFT Algorithm Kaggle Implementation 

```python
import cv2
import time
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
image1 = mpimg.imread("Images/0.jpg")
image2 = mpimg.imread("Images/0.jpg")

plt.figure()
plt.imshow(image1)

plt.figure()
plt.imshow(image2)
```

```python
#  Calculate the feature point extraction & generation description time
start = time.time()
sift = cv2.SIFT_create(nfeatures=10000)
#  Use SIFT to find key points and descriptors
kp1, des1 = sift.detectAndCompute(image1, None)
kp2, des2 = sift.detectAndCompute(image2, None)
end = time.time()
print("Feature Point Extraction & Generation description running time:%.2fseconds"%(end-start))

kp_image1 = cv2.drawKeypoints(image1, kp1, None)
kp_image2 = cv2.drawKeypoints(image2, kp2, None)

print(len(kp_image1))
print(len(kp_image2))

plt.figure()
plt.imshow(kp_image1)
#plt.savefig('kp_image1.png', dpi = 300)

plt.figure()
plt.imshow(kp_image2)
#plt.savefig('kp_image2.png', dpi = 300)
```

```python
#  View key points
print("Number of key points:", len(kp1))

for i in range(2):
    print("key point", i)
    print("data type:", type(kp1[i]))
    print("Key point coordinates:", kp1[i].pt)
    print("Neighborhood diameter:", kp1[i].size)
    print("direction:", kp1[i].angle)
    print("In the image pyramid group:", kp1[i].octave)
    print("================")

#  View Description
print("description shape:", des1.shape)
for i in range(2):
    print("Description", i)
    print(des1[i])
```

```python
ratio = 0.85

#  Calculate the matching time of matching points
start = time.time()

#  The k-nearest neighbors algorithm finds the K nearest data points in space and groups them together
matcher = cv2.BFMatcher()
raw_matches = matcher.knnMatch(des1, des2, k = 2)
good_matches = []
for m1, m2 in raw_matches:
    #  If the ratio of the closest point to the next closest point is greater than a given value, we keep the closest point and consider it and its matching point good_match
    if m1.distance < ratio * m2.distance:
        good_matches.append([m1])
end = time.time()
print("The matching point matches the running time:%.2fseconds"%(end-start))


matches = cv2.drawMatchesKnn(image1, kp1, image2, kp2, good_matches, None, flags = 2)

plt.figure()
plt.imshow(matches)
plt.savefig('matches.png', dpi = 300)
```

```python
print("The number of matching pairs:", len(good_matches))
for i in range(2):
    print("matching", i)
    print("data type:", type(good_matches[i][0]))
    print("The distance between descriptors:", good_matches[i][0].distance)
    print("Queries the index of descriptors in an image:", good_matches[i][0].queryIdx)
    print("Index of descriptors in the target image:", good_matches[i][0].trainIdx)
    print("================")
```

```python
import numpy as np

#  The homography matrix has eight parameters, and each corresponding pixel can generate two equations (one for x and one for y), so it takes four pixels to solve the homography matrix
if len(good_matches) > 4:
    #  Calculated matching time
    start = time.time()
    ptsA= np.float32([kp1[m[0].queryIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    ptsB = np.float32([kp2[m[0].trainIdx].pt for m in good_matches]).reshape(-1, 1, 2)
    ransacReprojThreshold = 4
    #  A homologous matrix can align one graph with another by rotation, transformation, etc
    H, status =cv2.findHomography(ptsA,ptsB,cv2.RANSAC,ransacReprojThreshold);
    imgOut = cv2.warpPerspective(image2, H, (image1.shape[1],image1.shape[0]),flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
    end = time.time()
    print("Matching running time:%.2fseconds"%(end-start))
    
    plt.figure()
    plt.imshow(image1)
    plt.figure()
    plt.imshow(image2)
    plt.figure()
    plt.imshow(imgOut)
    plt.savefig('imgOut.png', dpi = 300)
```

# CNN model ChatGPT code

```python
import tensorflow as tf
import numpy as np
import os
import cv2

# Load a pre-trained model (e.g., InceptionV3) for feature extraction
model = tf.keras.applications.InceptionV3(weights='imagenet', include_top=False)

# Define a function to extract features from an image
def extract_features(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (299, 299))  # Ensure the image size matches the model's requirements
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    img = np.expand_dims(img, axis=0)  # Add a batch dimension
    features = model.predict(img)
    return features
```

```python

# Directory containing images in your database
database_directory = 'path/to/database'

# Image to compare against the database
query_image_path = 'path/to/query_image.jpg'

# Extract features from the query image
query_features = extract_features(query_image_path)

# Initialize variables to keep track of the most similar image and its score
most_similar_image = None
best_score = float('-inf')

# Iterate through the images in the database directory
for filename in os.listdir(database_directory):
    if filename.endswith('.jpg'):
        database_image_path = os.path.join(database_directory, filename)
        database_features = extract_features(database_image_path)
        
        # Calculate the similarity score (e.g., cosine similarity)
        score = np.dot(query_features, database_features.T) / (np.linalg.norm(query_features) * np.linalg.norm(database_features))
        
        # Update the most similar image if a better match is found
        if score > best_score:
            best_score = score
            most_similar_image = database_image_path

print(f"The most similar image is: {most_similar_image} with a similarity score of {best_score}")
```
